```python
import numpy as np
import matplotlib.pyplot as plt
from RZutilpy.figure import bar, plot, default_img_set, colorpalettable
from scipy.stats import sem
import matplotlib.gridspec as gridspec
default_img_set()
import pingouin
```

```python
ryzcolor = colorpalettable('ryz1')
C0, C1, C0_alpha, C1_alpha = np.array(ryzcolor(0)), np.array(ryzcolor(1)),np.array(ryzcolor(0)), np.array(ryzcolor(1))
C0_alpha[-1] = 0.3
C1_alpha[-1] = 0.3
Clist = [np.array(ryzcolor(2)), np.array(ryzcolor(3)), np.array([0.121, 0.6039, 0.227]), np.array(ryzcolor(1))]
Calphalist = [np.array(ryzcolor(2)), np.array(ryzcolor(3)), np.array([0.121, 0.6039, 0.227, 1.0]), np.array(ryzcolor(1))]
Calphalist[0][-1]=0.3
Calphalist[1][-1]=0.3
Calphalist[2][-1]=0.3
Calphalist[3][-1]=0.3
```

```python
M1_Metrics = np.load('../Fig8_M1_Metrics.pkl', allow_pickle=True)
M2_Metrics = np.load('../Fig8_M2_Metrics.pkl', allow_pickle=True)
deltax=0.15

contrast = np.array([10, 15, 20, 25, 27, 28, 29, 31, 32, 33, 35, 40, 50, 60])
idx = np.arange(4, 10, 1)
n_condition = len(idx)
print(contrast[idx])
```

```python
decoding_pre = (M1_Metrics['decoding_pre'][idx]+M2_Metrics['decoding_pre'][idx])/2
decoding_post = (M1_Metrics['decoding_post'][idx]+M2_Metrics['decoding_post'][idx])/2
ACC_pre = ((M1_Metrics['ACC_pre'][idx]+ M2_Metrics['ACC_pre'][idx])/2).squeeze()
ACC_post = ((M1_Metrics['ACC_post'][idx]+ M2_Metrics['ACC_post'][idx])/2).squeeze()
LFI_pre = (M1_Metrics['LFI_pre'][idx]+ M2_Metrics['LFI_pre'][idx])/2
LFI_post = (M1_Metrics['LFI_post'][idx]+ M2_Metrics['LFI_post'][idx])/2
df_abs_pre = (M1_Metrics['df_abs_pre'][idx]+ M2_Metrics['df_abs_pre'][idx])/2
df_abs_post = (M1_Metrics['df_abs_post'][idx]+ M2_Metrics['df_abs_post'][idx])/2
FanoFactor_pre = ((np.median(np.array(M1_Metrics['FanoFactor_pre'])[idx],axis=2)+np.median(np.array(M2_Metrics['FanoFactor_pre'])[idx],axis=2))/2).squeeze()
FanoFactor_post = ((np.median(np.array(M1_Metrics['FanoFactor_post'])[idx],axis=2)+np.median(np.array(M2_Metrics['FanoFactor_post'])[idx],axis=2))/2).squeeze()
NoiseCorrelation_pre = ((np.median(np.array(M1_Metrics['correlation_pre'])[idx],axis=2)+np.median(np.array(M2_Metrics['correlation_pre'])[idx],axis=2))/2).squeeze()
NoiseCorrelation_post = ((np.median(np.array(M1_Metrics['correlation_post'])[idx],axis=2)+np.median(np.array(M2_Metrics['correlation_post'])[idx],axis=2))/2).squeeze()
variance_pre = ((np.median(np.array(M1_Metrics['variance_pre'])[idx],axis=1)+np.median(np.array(M2_Metrics['variance_pre'])[idx],axis=1))/2).squeeze()
variance_post = ((np.median(np.array(M1_Metrics['variance_post'])[idx],axis=1)+np.median(np.array(M2_Metrics['variance_post'])[idx],axis=1))/2).squeeze()
rotation = (M1_Metrics['rotation'][idx]+ M2_Metrics['rotation'][idx])/2
PCA_LFI = (np.array(M1_Metrics['PCA_LFI'][idx])+np.array(M2_Metrics['PCA_LFI'])[idx])/2
PCA_rotation = (np.array(M1_Metrics['PCA_rotation'])[idx][:,:20]+np.array(M2_Metrics['PCA_rotation'])[idx])/2
PC_abs_pre = (np.array(M1_Metrics['PC_abs_pre'])[idx][:,:20]+np.array(M2_Metrics['PC_abs_pre'])[idx])/2
PC_abs_post = (np.array(M1_Metrics['PC_abs_post'])[idx][:,:20]+np.array(M2_Metrics['PC_abs_post'])[idx])/2
```

```python
from scipy.stats import ttest_rel
print(ttest_rel(ACC_pre, ACC_post, alternative='less'))
print(ttest_rel(decoding_pre, decoding_post, alternative='less'))
print(ttest_rel(LFI_pre, LFI_post, alternative='less'))
print(ttest_rel(FanoFactor_pre, FanoFactor_post, alternative='greater'))
print(ttest_rel(NoiseCorrelation_pre, NoiseCorrelation_post, alternative='greater'))
print(ttest_rel(df_abs_pre, df_abs_post, alternative='less'))
print(ttest_rel(variance_pre, variance_post, alternative='greater'))

# Also calculate the effect size for each metric
from pingouin import compute_effsize
print(compute_effsize(ACC_pre, ACC_post, paired=True))
print(compute_effsize(decoding_pre, decoding_post, paired=True))
print(compute_effsize(LFI_pre, LFI_post, paired=True))
print(compute_effsize(FanoFactor_pre, FanoFactor_post, paired=True))
print(compute_effsize(NoiseCorrelation_pre, NoiseCorrelation_post, paired=True))
print(compute_effsize(df_abs_pre, df_abs_post, paired=True))
print(compute_effsize(variance_pre, variance_post, paired=True))
```

```python
fig, ax=plt.subplots(1,1, figsize=(2,3))
plt.sca(ax)
plot(np.array([1-deltax, 1+deltax]), np.vstack((ACC_pre,ACC_post)), fmt='-', color='lightgray')
plot([1-deltax]*n_condition, ACC_pre, fmt='o', markerfacecolor=C0_alpha, markeredgecolor=C0)
plot([1+deltax]*n_condition, ACC_post, fmt='o', markerfacecolor=C1_alpha, markeredgecolor=C1)
bar([1], np.vstack((ACC_pre.mean(axis=0), ACC_post.mean(axis=0))).T, 
yerr = np.vstack((sem(ACC_pre,axis=0), sem(ACC_post,axis=0))).T
,ecolor='k', width=0.2, bargap=0,facecolor=[C0_alpha,C1_alpha],edgecolor=[ryzcolor(0),ryzcolor(1)],zorder=10)
ax.set_ylim([0.50, 0.80])
ax.set_yticks([0.5, 0.6, 0.7, 0.8])
ax.set_xlabel(" ")
ax.set_ylabel("Accuracy")
ax.set_xticks([1])
ax.set_xticklabels([' '])
plt.savefig('Fig8_acc.pdf')
```
